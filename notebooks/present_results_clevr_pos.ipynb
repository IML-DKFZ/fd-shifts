## <br><br><br> Scatterplot (te_kh_po_tr = telbo_khot_poe_tracking)

```python
fig, ax = plt.subplots()
df300['exp_name'] = ["300_" + i for i in df300['exp_name']]
df_all = pd.concat([df, df300])
df_all.plot('correctness_total', 'coverage_total', kind='scatter', ax=ax, figsize=(40, 40))

for k, v in df_all.iterrows():
    ax.annotate(('_').join([i[:2] for i in v['exp_name'].split('_')]), v[['correctness_total', 'coverage_total']], fontsize=12)
    
fig.canvas.draw()
```
